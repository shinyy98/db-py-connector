```python
# ========== 方法2: 使用预设配置 ==========
print("\n" + "=" * 50)
print("方法2: 使用预设配置连接")
print("=" * 50)

from DatabasePyConn import MySQLDB
import pandas as pd

# 预设配置在 DatabasePyConn.py 中的 MySQLDB.DB_CONFIGS 定义
try:
    with MySQLDB.from_config_key('winddb') as db:
        print("✅ 使用预设配置连接成功！")
        result = db.query("SELECT 1 as test")
        print(f"测试查询结果: {result.iloc[0]['test']}")
except Exception as e:
    print(f"❌ 连接失败: {e}")

```
